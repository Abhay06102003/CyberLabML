```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def normalization(X):
  X1 = []
  for x in X.T[:]:
    x= (x - x.mean())/(np.std(x))
    X1.append(x)
  X1 = np.array(X1)
  X1 = X1.T
  return X1

def multi_lin(X,y,lr,no_iter):
  X = normalization(X)
  m,n = X.shape
  w = np.zeros((n,1))
  b = 0
  costs = []
  lst = [i for i in range(no_iter)]
 
  
  for i in range(no_iter):
    h_x = np.dot(X,w) +b
    cost = np.sum((h_x - y)**2)*(1/(2*m))


    w_grad = (1/m)*np.dot(X.T,(h_x -y))
    b_grad = (1/m)*np.sum(h_x-y)

    #Update
    w = w - (lr*w_grad)
    b = b - (lr*b_grad)
    costs.append(cost)
    if i % (no_iter/10) == 0:
      print('cost after ',i,' iteration is ',cost)
  plt.plot(lst,costs)
  plt.xlabel('No.of iter')
  plt.ylabel('cost ')
  plt.show()
  return w,b,h_x
def testing(w,b,Xtest):
  ytest = np.dot(Xtest,w) + b
  return ytest
```

```python
df = pd.read_csv('Linear_train.csv',header=0)
df
```

```python
X = np.array([df['A'],df['B'],df['C'],df['D'],df['E'],df['F'],df['G'],df['H'],df['I'],df['J'],df['K'],df['L'],df['M'],df['N'],df['O'],df['P'],df['Q'],df['R'],df['S'],df['T']])
X = X.T
X = normalization(normalization(X))
```

```python
y = np.array(df['label'])
y = y.reshape(50000,1)
y.shape
```

```python
w,b,h_x = multi_lin(X,y,0.001,50000)
```

```python
Xtest = np.array([df1['A'],df1['B'],df1['C'],df1['D'],df1['E'],df1['F'],df1['G'],df1['H'],df1['I'],df1['J'],df1['K'],df1['L'],df1['M'],df1['N'],df1['O'],df1['P'],df1['Q'],df1['R'],df1['S'],df1['T']])
Xtest = Xtest.T
Xtest = normalization(Xtest)
Xtest.shape
```

```python
ypred = testing(w,b,Xtest)
ypred
```
