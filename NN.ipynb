```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def sig(x):
  return 1/(1+np.exp(-x))
def d(x):
  a = sig(x)
  return a*(1-a)
def normalization(X2):
  X1 = []
  for x in X2.T[:]:
    x = (x - x.mean())/(np.std(x)+1)
    X1.append(x)
  X1 = np.array(X1)
  X1 = X1.T
  return X1
def NN(X,y,lr,no_iter,hid_units):
  X = normalization(X)
  m,n = X.shape
  w1 = np.random.rand(n,hid_units)
  w2 = np.random.rand(hid_units,y.shape[1])
  b1 = np.random.rand(hid_units,1)
  costs = []
  lst = [ i for i in range(no_iter)]
  for i in range(no_iter):
    #feedforward
    a1 = sig((np.dot(X,w1)).T+b1)
    out = sig(np.dot(a1.T,w2))
    #backprop
    e1 = 2*(y-out)*d(out)
    dw1 = (1/m)*np.dot(X.T,((np.dot(e1,w2.T))*(d(a1)).T))
    dw2 = (1/m)*np.dot(a1,e1)
    db1 = (1/m)*np.sum(((np.dot(e1,w2.T)).T)*d(a1),axis=1)
    #print(db1.shape)
    db1 = db1.reshape(hid_units,1)
    #update
    w1 = w1 + lr*dw1
    w2 = w2 + lr*dw2
    b1 = b1 + lr*db1
    #cost
    cost = (1/(2*m))*np.sum((y-out)**2)
    costs.append(cost)
    if i%(no_iter/10)==0:
      print('cost after ',i,' iterartions is ',cost)
  
    
  plt.plot(lst,costs,'y')
  plt.show()
  return w1,w2,b1
def test(w1,w2,b1,Xtest):
  a1 = sig((np.dot(Xtest,w1)).T+b1)
  output = sig(np.dot(a1.T,w2))
  return output
```

```python
X = []
for i in range(88799):
  a = dftrainX.loc[i]
  X.append(a)
X = np.array(X)
X = normalization(X)
X.shape
```

```python
W1 =[]
W2=[]
B=[]
for i in range(1,27):
  y = []
  for j in range(X.shape[0]):
    y.append(dftrainY.loc[j,[str(i)]])
  y = np.array(y)
  w1,w2,b1 = NN(X,y,0.7,200,26)
  W1.append(w1)
  W2.append(w2)
  B.append(b1)
```
