```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def sig(x):
  return 1/(1+np.exp(-x))
def d(x):
  a = sig(x)
  return a*(1-a)
def normalization(X2):
  X1 = []
  for x in X2.T[:]:
    x = (x - x.mean())/(np.std(x)+1)
    X1.append(x)
  X1 = np.array(X1)
  X1 = X1.T
  return X1
def NN(X,y,lr,no_iter):
  X = normalization(X)
  m,n = X.shape
  w = np.zeros((n,1))
  b = 0
  lst = [i for i in range(no_iter)]
  costs = []
  for i in range(no_iter):
    out = sig(np.dot(X,w)+b)
    error = y - out
    dw = (1/m)*np.dot(X.T,(d(out)*error))
    db = (1/m)*np.sum((d(out)*error))
    w += lr*dw
    b += lr*db
    cost = (1/(2*m))*np.sum((y-out)**2)
    costs.append(cost)
    if i%(no_iter/10)==0:
      print('cost after ',i,' iterartions is ',cost)
  plt.plot(lst,costs,'y')
  plt.show()
  return w,b
def testing(w,b,Xtest):
  return sig(np.dot(Xtest,w)+b)
```

```python
dftrain = pd.read_csv('emnist-letters-train.csv')
dftrainX  = pd.read_csv('emnist-letters-train.csv')
dftrainX.drop(columns = ['23'],inplace = True)
dftrainY = pd.read_csv('emnist-letters-train.csv',header = 0)
dftrainY.drop(columns = ['23'],inplace = True)
for i in range(1,27):
  a = list(dftrain['23'] == i)
  for j in range(len(a)):
    a[j] = int(a[j])
  dftrainY[str(i)] = a
```

```python
X = []
for i in range(88799):
  a = dftrainX.loc[i]
  X.append(a)
X = np.array(X)
X = normalization(X)
X.shape
```

```python
W =[]
B=[]
for i in range(1,27):
  y = []
  for j in range(X.shape[0]):
    y.append(dftrainY.loc[j,[str(i)]])
  y = np.array(y)
  w,b = NN(X,y,1.5,1000)
  W.append(w)
  B.append(b)
```

```python
dftest = pd.read_csv('emnist-letters-test.csv')
dftestX = pd.read_csv('emnist-letters-test.csv')
dftestX = dftestX.drop(columns = ['1'])
```

```python
Xtest = []
for i in range(14799):
  a = dftestX.loc[i]
  Xtest.append(a)
Xtest = np.array(Xtest)
Xtest = normalization(Xtest)
Xtest.shape
ytest = np.array(dftest['1'])
```

```python
ypreds = []
W = np.array(W)
B = np.array(B)
for i in range(26):
  ypred = testing(W[i],B[i],Xtest)
  ypreds.append(ypred)
ypreds = np.array(ypreds)
ypreds = ypreds.reshape(26,14799)
outlst =(ypreds).argmax(axis = 0)+1
outlst = np.array(outlst)
outs = pd.DataFrame(outlst)
outs
```

```python
correct = 0
for i in range(Xtest.shape[0]):
  if outlst[i] == ytest[i]:
    correct += 1
(correct/14799)*100
```
